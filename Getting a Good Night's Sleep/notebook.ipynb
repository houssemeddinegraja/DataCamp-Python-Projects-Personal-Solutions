Your client, SleepInc, has shared anonymized sleep data from their hot new sleep tracking app SleepScope. As their data science consultant, your mission is to analyze the lifestyle survey data with Python to discover relationships between exercise, gender, occupation, and sleep quality. See if you can identify patterns leading to insights on sleep quality.

## 💾 The data: sleep_health_data.csv

SleepInc has provided you with an anonymized dataset of sleep and lifestyle metrics for 374 individuals. This dataset contains average values for each person calculated over the past six months. The data is saved as `sleep_health_data.csv`.

The dataset includes 13 columns covering sleep duration, quality, disorders, exercise, stress, diet, demographics, and other factors related to sleep health. 

| Column | Description |
|---------|----------------------------------------|  
| `Person ID` | An identifier for each individual. |
| `Gender` | The gender of the person (Male/Female). |  
| `Age` | The age of the person in years. |
| `Occupation` | The occupation or profession of the person. |
| `Sleep Duration (hours)` | The average number of hours the person sleeps per day. |
| `Quality of Sleep (scale: 1-10)` | A subjective rating of the quality of sleep, ranging from 1 to 10. |
| `Physical Activity Level (minutes/day)` | The average number of minutes the person engages in physical activity daily. |  
| `Stress Level (scale: 1-10)` | A subjective rating of the stress level experienced by the person, ranging from 1 to 10. |
| `BMI Category` | The BMI category of the person (e.g., Underweight, Normal, Overweight). |
| `Blood Pressure (systolic/diastolic)` | The average blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure. |
| `Heart Rate (bpm)` | The average resting heart rate of the person in beats per minute. |
| `Daily Steps` | The average number of steps the person takes per day. |
| `Sleep Disorder` | The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea). |

```python
import pandas as pd

#Converting the given csv file into a DataFrame
sleep_health_data_df = pd.read_csv("sleep_health_data.csv")
```

```python
#First Task:

#Grouping the occupations and calculting the mean sleep duration (using a pivot table)
sleep_duration_means=sleep_health_data_df.pivot_table(values="Sleep Duration",index="Occupation")

#Sorting occupations according to mean sleep durations
sleep_duration_means_srtd=sleep_duration_means.sort_values("Sleep Duration")

#Getting the occupation with the lowest mean sleep duration
lowest_sleep_occ=sleep_duration_means_srtd.index[0]
print(lowest_sleep_occ)
```

```python
#Second Task:
#Following approximately the same instructions as the previous task, but with sleep quality

#Grouping the occupations and calculting the mean sleep quality (using a pivot table)
sleep_quality_means=sleep_health_data_df.pivot_table(values="Quality of Sleep",index="Occupation")

#Sorting occupations according to mean sleep qualities
sleep_quality_means_srtd=sleep_quality_means.sort_values("Quality of Sleep")

#Getting the occupation with the lowest mean sleep quality
lowest_sleep_quality_occ=sleep_quality_means_srtd.index[0]
print(lowest_sleep_quality_occ,"\n")

#Checking whether the occupation with the lowest mean sleep duration is the occupation with the lowest mean sleep quality
same_occ=True
if lowest_sleep_occ==lowest_sleep_quality_occ:
    same_occ=True
    print("--> The occupation that had the lowest average sleep time is the same that had the lowest average sleep quality; ",lowest_sleep_occ)
else:
    same_occ=False
    print("--> The occupation that had the lowest average sleep time (",lowest_sleep_occ,") is not the same occupation that had the lowest average sleep quality (",lowest_sleep_quality_occ,").")
```

```python
#Third Task:

#Grouping the sleep disorders according to BMIs
bmi_disorder=sleep_health_data_df[["BMI Category","Sleep Disorder"]]

#Calculating the ratio of users diagnosed with Insomnia in each BMI category

#Normal
total_normal=len(bmi_disorder[bmi_disorder["BMI Category"]=="Normal"])
normal_insomnia=bmi_disorder[(bmi_disorder["Sleep Disorder"]=="Insomnia")&(bmi_disorder["BMI Category"]=="Normal")]
normal_insom_cnt=len(normal_insomnia)
normal_ratio=round(normal_insom_cnt/total_normal,2)

#Overweight
total_ovw=len(bmi_disorder[bmi_disorder["BMI Category"]=="Overweight"])
ovw_insomnia=bmi_disorder[(bmi_disorder["Sleep Disorder"]=="Insomnia")&(bmi_disorder["BMI Category"]=="Overweight")]
ovw_insom_cnt=len(ovw_insomnia)
ovw_ratio=round(ovw_insom_cnt/total_ovw,2)

#Obese
total_obese=len(bmi_disorder[bmi_disorder["BMI Category"]=="Obese"])
obese_insomnia=bmi_disorder[(bmi_disorder["Sleep Disorder"]=="Insomnia")&(bmi_disorder["BMI Category"]=="Obese")]
obese_insom_cnt=len(obese_insomnia)
obese_ratio=round(obese_insom_cnt/total_obese,2)
bmi_insomnia_ratios={ "Normal" : normal_ratio ,
                    "Overweight" : ovw_ratio ,
                    "Obese" : obese_ratio 
                    }
```

```python
#Final Printing
print("Here's the ratio of app users diagnosed with Insomnia according to BMI categories:")
print(bmi_insomnia_ratios)
```
