The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Start coding here!

#___________________________________________________________________________________

# Loading the dataset
nobel_df = pd.read_csv("data/nobel.csv")

# Getting the most awarded gender
gender_cnt = nobel_df["sex"].value_counts()
top_gender = gender_cnt.index[0]
#print(top_gender)

# Getting the most awarded birth country
brthcntry_cnt = nobel_df["birth_country"].value_counts()
top_country = brthcntry_cnt.index[0]
#print(top_country)

# Calculating the ratio of US-born winners to total all-categories winners
nobel_df["usborn_winner"] = nobel_df["birth_country"]=="United States of America"
nobel_df["decade"] = (nobel_df["year"]//10)*10
usborn_winners = (nobel_df.groupby("decade",as_index=False)["usborn_winner"].mean().sort_values("usborn_winner",ascending=False))
#print(usborn_winners)
max_decade_usa = usborn_winners["decade"].iloc[0]
#print(max_decade_usa)

# Finding the Decade:Category with the highest proportion of females
nobel_df["female_winner"]=nobel_df["sex"]=="Female"
fem_grpd = nobel_df.groupby(["decade","category"],as_index=False)["female_winner"].mean().sort_values("female_winner",ascending=False)
max_female_dict = { fem_grpd["decade"].iloc[0]:fem_grpd["category"].iloc[0] }
#print(max_female_dict)

# Finding the first woman to win a Nobel prize
female_df = nobel_df[nobel_df["sex"]=="Female"]
female_datesrtd = female_df.sort_values("year")
first_woman_name = female_datesrtd["full_name"].iloc[0]
first_woman_category = female_datesrtd["category"].iloc[0]
#print(first_woman_name,first_woman_category)

# Finding repeat winners
nobel_cnts = (nobel_df["full_name"].value_counts()).to_frame()
nobel_repeats = nobel_cnts[nobel_cnts["full_name"]>1]
repeat_list = []
for winner in nobel_repeats.index:
    repeat_list.append(winner)
print(repeat_list)
```
